## 3. Análisis de Marginalización: Porcentaje de Municipios por Estado

La siguiente gráfica muestra el porcentaje de municipios por estado con grado de marginación 
- "muy bajo"
- "bajo"
- "medio"
- "alto"
- "muy alto"

Esto permite realizar un análisis de la marginación y desigualdad por estados.

```python
# Creamos un dataframe con la información relevante para la gráfica
distribucion_df = marginacion_df.groupby(['NOM_ENT','GM_2020'])[['NOM_MUN']].count().reset_index()

# Calculamos el total de municipios por estados para poder calcular el porcentaje relativo
group_totals = distribucion_df.groupby('NOM_ENT')['NOM_MUN'].transform('sum')

# Calcula el porcentaje relativo, esto es, cantidad de municipios en cada categiría de índice sobre el total de municipios
distribucion_df['Porcentaje'] = (distribucion_df['NOM_MUN'] / group_totals) * 100

distribucion_df.head()
```

```python
#Crear la gráfica
fig = px.bar(distribucion_df, x='NOM_ENT', y='Porcentaje', color='GM_2020', hover_name='GM_2020',
             title='Porcentaje de municipios por estado por grado de marginación',
             category_orders={'GM_2020': ['Muy alto', 'Alto', 'Medio', 'Bajo', 'Muy bajo']})
# Editar título del eje x
fig.update_xaxes(title_text='Estado')
# Editar título del eje y
fig.update_yaxes(title_text='Porcentaje de municipios')
# Editar título de la leyenda
fig.update_layout(legend_title_text='Grado de marginación')
# Editar tooltips
fig.update_traces(hovertemplate='<b>%{x}</b><br>' +
                                'Grado de marginación: %{hovertext}<br>' +
                                'Porcentaje de municipios: %{y:.1f}%')
# Personalizar el acomodo de las etiquetas del eje X y la altura de la gráfica
fig.update_layout(
    xaxis=dict(tickangle=315),
    height = 600
)
# Guardamos la gráfica como una imagen estática en formato png
pio.write_image(fig, 'grafica1.png')

# Imprimimos la gráfica
fig.show()
```

## 4. Análisis de Marginación: Porcentaje de Población por Estado

```python
# Creamos el dataframe con la información agrupada que se requiere para las gráficas
distribucion_poblacion_df = marginacion_df.groupby(['NOM_ENT','GM_2020'])[['POB_TOT']].sum().reset_index()

# Calculamos el total de población por estado
total_por_estado = distribucion_poblacion_df.groupby('NOM_ENT')['POB_TOT'].transform('sum')

# Calculamos el porcentaje relativo de la distribución de la población
distribucion_poblacion_df['Porcentaje'] = (distribucion_poblacion_df['POB_TOT'] / total_por_estado) * 100

# Renombramos las columnas
distribucion_poblacion_df = distribucion_poblacion_df.rename(columns={'NOM_ENT':'Estado', 'GM_2020': 'Grado de marginación', 'POB_TOT':'Población'})
distribucion_poblacion_df.head()
```

```python
# Creamos la gráfica dinámica en plotly
fig = px.bar(distribucion_poblacion_df, x='Estado', y='Porcentaje', color='Grado de marginación', hover_name='Grado de marginación',
             title='Porcentaje de población por estado por grado de marginación',
             category_orders={'Grado de marginación': ['Muy alto', 'Alto', 'Medio', 'Bajo', 'Muy bajo']})
# Editamos título del eje y
fig.update_yaxes(title_text='Porcentaje de municipios')
# Editamos tooltips
fig.update_traces(hovertemplate='<b>%{x}</b><br>' +
                                'Grado de marginación: %{hovertext}<br>' +
                                'Porcentaje de población: %{y:.1f}%')
# Personalizamos la ubicación de las etiquetas del eje X y el tamaño de la gráfica
fig.update_layout(
    xaxis=dict(tickangle=315),
    height = 600
)
# Guardamos la gráfica en formato jpg
pio.write_image(fig, 'grafica2.jpg')
fig.show()
```

## 7. Análisis de correlación 
<t align="justify">

Para poder identificar las variables que tienen una mayor correlación con el porcentaje de analfabetismo en personas mayores a 15 años debemos calcular una matriz de correlación que incluya todas las variables en cuestión.

Recordemos que la matriz de correlación es una herramienta estadística que permite analizar la relación entre variables. Se utiliza para determinar la fuerza y la dirección de la asociación entre pares de variables en un conjunto de datos. Esta matriz se representa como una tabla cuadrada donde cada celda contiene un valor que indica el grado de correlación entre dos variables.

La correlación es una medida estadística que cuantifica la relación lineal entre dos variables. Puede tomar valores en el rango de -1 a 1. Un valor de 1 indica una correlación positiva perfecta, lo que significa que a medida que una variable aumenta, la otra también lo hace de manera proporcional. Por otro lado, un valor de -1 indica una correlación negativa perfecta, lo que significa que a medida que una variable aumenta, la otra disminuye en forma proporcional. Un valor de 0 indica una ausencia de correlación, es decir, no hay relación lineal entre las variables.

Para calcular cada entrada de la matriz de correlación, se utiliza un coeficiente de correlación, el más comúnmente utilizado es el **coeficiente de correlación de Pearson**. Este coeficiente se calcula dividiendo la covarianza de las dos variables por el producto de las desviaciones estándar de cada variable. La covarianza mide cómo varían conjuntamente las dos variables, mientras que las desviaciones estándar miden la dispersión de cada variable por separado. El coeficiente de correlación de Pearson varía entre -1 y 1, donde un valor cercano a 1 o -1 indica una correlación fuerte, y un valor cercano a 0 indica una correlación débil o nula.

La matriz de correlación es una herramienta fundamental en el análisis de datos, ya que permite identificar patrones, dependencias y tendencias entre variables. Al examinar la matriz de correlación, es posible determinar qué variables están fuertemente relacionadas entre sí y cuáles pueden considerarse independientes. Esto proporciona información valiosa para el análisis que deseamos realizar.

```python
# Seleccionamos las variables de interés
marginacion_num_df = marginacion_df[['POB_TOT', 'ANALF', 'SBASC',
       'OVSDE', 'OVSEE', 'OVSAE', 'OVPT', 'VHAC', 'PL.5000', 'PO2SM',
       'IM_2020']]
```

```python
# Calculamos la matriz de correlación

matriz_corr = marginacion_num_df.corr()
matriz_corr
```

A partir de esta matriz es muy dificil sacar conclusiones, incluso distinguir el valor mayor es complicado. Por lo que, para realizar el análisis de correlación graficaremos la matriz de correlación en un heatmap.

```python
# Creamos un heatmap para poder visualizar los valores de los coeficientes de correlación
fig = px.imshow(matriz_corr.values,
                x=matriz_corr.columns,
                y=matriz_corr.index,
                title='Matriz de correlación')

# Personalizamos el tamaño del heatmap
fig.update_layout(width=600,height=600)

# Graficamos
fig.show()
```

<t align="justify">

A partir de este heatmap, podemos observar las principales variables correlacionadas con el porcentaje de analfabetismo en la población:

1. **Índice de marginación:** Como era de esperar, el índice de marginación muestra una correlación significativa con el porcentaje de analfabetismo. Esto se debe a que el índice de marginación incluye el porcentaje de analfabetismo como uno de sus indicadores, por lo que existe una relación directa entre ambas variables.

2. **Porcentaje de población de 15 años o más sin educación básica (SBASC):** Existe una correlación alta y positiva de 0.77 entre esta variable y el porcentaje de analfabetismo. 3. **Porcentaje de viviendas particulares con hacinamiento (VHAC):** Se observa una correlación mediana y positiva de 0.65 entre el hacinamiento y el analfabetismo. El hacinamiento se caracteriza por la falta de suficiente espacio habitacional adecuado, lo cual dificulta las condiciones propicias para el estudio y puede contribuir a un mayor nivel de analfabetismo.

4. **Porcentaje de ocupantes en viviendas particulares habitadas con piso de tierra (OVPT):** Con un coeficiente de correlación de 0.64, se evidencia una relación positiva y mediana entre esta variable y el analfabetismo. Esta característica también está relacionada con condiciones de pobreza, y su presencia puede estar asociada a un mayor nivel de analfabetismo.

5. **Porcentaje de población ocupada con ingresos de hasta 2 salarios mínimos (PO2SM):** Se identifica una correlación positiva de magnitud mediana (coeficiente de 0.62) entre esta variable y el analfabetismo. Aunque la correlación no es fuerte, indica que existe una relación directa entre la proporción de población con ingresos bajos y el porcentaje de analfabetismo. Esta variable también refleja condiciones económicas desfavorables que pueden dificultar el acceso a la educación.

Estas correlaciones resaltan la importancia de abordar la educación y la pobreza en conjunto, ya que están estrechamente relacionadas. Implementar políticas y programas que fomenten la educación inclusiva y mejoren las condiciones socioeconómicas puede ser fundamental para reducir el nivel de analfabetismo y promover un desarrollo equitativo en la sociedad.

## 8. Propuesta de indicadores: Índice de Marginación de Vivienda

<t align="justify">

El Índice de Marginación en Vivienda es un indicador que se crea a partir de la combinación de diversas variables relacionadas con las condiciones de vivienda en una determinada área geográfica. Este índice permite evaluar la calidad de las condiciones de vivienda y reflejar el grado de marginación en este aspecto.

El índice se construye utilizando los siguientes porcentajes de viviendas:

- **Porcentaje de viviendas sin drenaje ni excusado (OVSDE):** Este indicador mide la falta de acceso a servicios básicos de saneamiento en las viviendas.

- **Porcentaje de viviendas sin energía eléctrica (OVSEE):** Este indicador muestra la falta de acceso a la electricidad en las viviendas.

- **Porcentaje de viviendas sin agua entubada (OVSAE):** Este indicador refleja la falta de acceso a agua potable a través de una red de distribución en las viviendas.

- **Porcentaje de viviendas con piso de tierra (OVPT):** Este indicador indica la ausencia de un piso adecuado en las viviendas, lo cual puede ser un indicio de condiciones precarias de vivienda.

- **Porcentaje de viviendas con hacinamiento (VHAC):** Este indicador muestra la proporción de viviendas donde existe una alta densidad de ocupantes en relación con el espacio disponible, lo cual puede ser un indicador de condiciones de hacinamiento.

Para calcular el Índice de Marginación en Vivienda, se pueden asignar pesos a cada una de estas variables según su importancia relativa en la evaluación de las condiciones de vivienda. Estos pesos pueden ser determinados mediante métodos estadísticos o mediante la opinión de expertos en el tema. Una vez asignados los pesos, se realiza una combinación ponderada de los porcentajes de viviendas en cada variable para obtener el índice.

Este índice proporciona una medida agregada de la marginación en el ámbito de la vivienda en una determinada área geográfica, y puede ser utilizado como una herramienta para identificar áreas con condiciones de vivienda precarias y orientar la implementación de políticas y programas de mejoramiento de vivienda en esas áreas.

Los pesos para este indicador los asignaré de manera proporcional a la correlación que existe con el nivel de analfabetismo. Este indicador podría resultar de utilidad para la creación de estrategias y políticas públicas orientadas a disminuir los niveles de analfabetismo en los sectores que presentan niveles muy altos de marginación de vivienda.

```python
# Seleccionamos los coeficientes de correlación
pesos = matriz_corr.loc['ANALF', ['OVSDE', 'OVSEE', 'OVSAE', 'OVPT', 'VHAC']]

# Normalizamos los pesos para que sumen 1 
pesos = np.divide(pesos, pesos.sum())
pesos
```

```python
# Creamos el nuevo dataframe con este indicador
marginacion_df_nuevo = marginacion_df.loc[:,['CVE_ENT', 'NOM_ENT', 'CVE_MUN', 'NOM_MUN', 'POB_TOT', 'ANALF', 'PL.5000', 'PO2SM',
       'IM_2020', 'GM_2020', 'IMN_2020']]
marginacion_df_nuevo.loc[:,'IMV_2020'] = (pesos[0]*marginacion_df['OVSDE'] + 
                                          pesos[1]*marginacion_df['OVSEE'] + 
                                          pesos[2]*marginacion_df['OVSAE'] + 
                                          pesos[3]*marginacion_df['OVPT'] +
                                          pesos[4]*marginacion_df['VHAC']                             
)
marginacion_df_nuevo.head()
```

```python
marginacion_df_nuevo.loc[:,'IMV_2020'].describe()
```

```python
# Veremos la distribución del índice de marginación de vivienda
df = px.data.tips()
fig = px.histogram(marginacion_df_nuevo, x="IMV_2020", nbins=50, title='Distribución del Índice de Marginación de Vivienda municipal',)
# Editar título del eje x
fig.update_xaxes(title_text='Valores del Índice de Marginación de Vivienda (IMV_2020)')
# Editar título del eje y
fig.update_yaxes(title_text='Cantidad de municipios')

fig.show()
```

Estos resultados resaltan la necesidad de implementar políticas y acciones enfocadas en reducir la marginación y promover la equidad en el acceso a la educación, vivienda adecuada y oportunidades económicas. El desarrollo de indicadores como el Índice de Marginación en Vivienda propuesto proporciona una herramienta valiosa para evaluar la calidad de vida de las comunidades y orientar los esfuerzos de intervención.
